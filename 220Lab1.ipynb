```python
#task2
def k_l_rotate(A, k):
  l = len(A)
  c = arr.array("i", [0]*k)
  d = arr.array("i", [0]*(l-k))
  for i in range(k):
    c[i]= A[i]
  j = 0
  for i in range(k, l):
    d[j]= A[i]
    j += 1
  i = 0
  for x in d:
    A[i] = x
    i += 1
  for x in c:
    A[i] = x
    i += 1
A = [10,20,30,40,50,60]
k_l_rotate(A, 3)
print(A)
```

```python
#Task3
def remove(A, sz, idx):
  for i in range(idx, sz-1):
    A[i] = A[i+1]
  A[sz-1] = 0
source=[10,20,30,40,50,0,0]
remove(source, 5, 2)
print(source)
```

```python
#task5
def splitting(A):
  l=len(A)
  for i in range(1,l):
    sum_lft=0
    for j in range(i):
      sum_lft+=A[j]
    sum_rgt=0
    for j in range(i,l):
      sum_rgt+=A[j]
    if sum_lft==sum_rgt:
      return True
  return False
A=arr.array('i',[1,1,1,2,1])
print(splitting(A))
B=arr.array('i',[2,1,1,2,1])
print(splitting(B))
C=arr.array('i',[10,3,10,2,1])
print(splitting(C))
```

```python
#task6
def Array_Series(N):
  C=arr.array('i',[0]*(N*N))
  for i in range(N):
    pos=(i+1)*N-1
    for j in range(i+1):
      C[pos]=j+1
      pos-=1
  return C
N=int(input())
A=Array_Series(N)
print(A)
```

```python
#task7
def Bunch(A):
  Mx=0
  l=len(A)
  i=0
  while i<l:
    cnt=0
    j=i
    while j<l and A[j]==A[i]:
      cnt+=1
      j+=1
    Mx=max(Mx,cnt)
    i=j
  if Mx<2:
    print('There is no Bunch')
  else :
    return Mx
A=arr.array('i',[1,2,2,3,4,4,4])
print(Bunch(A))
```

```python
#circular array1
def isPalindrome(A,sz,st):
  l=len(A)
  rear=(st+sz-1)%l
  i=st
  j=rear
  while sz>1:
    if A[i]==A[j]:
      i+=1
      j-=1
      if i==l:
        i=0
      if j==-1:
        j=l-1
      sz-=2
    else:
      return False
  return True
A=arr.array('i',[20,10,0,0,0,10,20,30])
st=5
sz=5
print(isPalindrome(A,sz,st))
```

```python
#circular array2
def intersection(A,sza,sta,B,szb,stb):
  temp=arr.array('i',[0]*len(A))
  temp_now=0
  total_size_A=len(A)
  total_size_B=len(B)
  A_now=sta
  for i in range(sza):
    B_now=stb
    for j in range(szb):
      if A[A_now]==B[B_now]:
        temp[temp_now]=A[A_now]
        temp_now+=1
        #print(A[A_now])
        break
      else :
        B_now+=1
        if B_now==total_size_B:
          B_now=0
    A_now+=1
    if A_now==total_size_A:
      A_now=0
  ret=arr.array('i',[0]*temp_now)
  for i in range(temp_now):
    ret[i]=temp[i]
  return ret
A=arr.array('i',[40,50,0,0,0,10,20,30])
sta=5
sza=5
B=arr.array('i',[10,20,5,0,0,0,0,0,5,40,15,25])
stb=8
szb=7
inter=intersection(A,sza,sta,B,szb,stb)
print(inter)
```

```python
class queue_cir_array:
   def __init__(self,sz):
     self.arr=np.array([0]*sz)
     self.front=None
     self.rear=-1
     self.length=sz
     self.sz=0
   def enqueue(self,val):
     if ( (self.rear+1)%self.length == self.front ):
       print('Overflow')
       return
     self.rear=(self.rear+1)%self.length
     self.arr[self.rear]=val
     self.sz=self.sz+1
     if self.sz==1:
       self.front=self.rear
     return
   def deque(self):
     if self.sz==0:
       return 'Underflow'
     val=self.arr[self.front]
     self.front=(self.front+1)%self.length
     self.sz=self.sz-1
     return val
   def peek(self):
     if self.sz==0:
       return 'Underflow'
     return self.arr[self.front]
   def size(self):
     return self.sz
```

```python
class queue_linked_list:
  def __init__(self):
    self.front=None
    self.rear=None
    self.size=0
  def enqueue(self,val):
    newNode=Node(val)
    if self.front == None:
      self.front=newNode
    else:
      self.rear.next=newNode

    self.rear=newNode
    self.size=self.size+1

  def deque(self):
    if self.size == 0:
      return 'Underflow'
    ret = self.front
    self.front=self.front.next
    self.size=self.size-1
    return ret.val
  def peek(self):
    if self.size == 0:
      return 'Underflow'
    return self.front.val
  def Size(self):
    return self.size
```

```python
#Priority Queue

#Sorted

class Sorted_PQ_Implementation:
  def __init__(self,n):
    self.pq=np.array([0]*n)
    self.top=-1
  def Insert(self,val):

    if self.top == -1:
      self.pq[0]=val
      self.top+=1
      return

    i=self.top
    while i>=0 and self.pq[i] < val:
      self.pq[i+1]=self.pq[i]
      i-=1
    
    self.pq[i+1]=val
    self.top+=1

  def Peek(self):
    if self.top == -1:
      return 'Underflow'
    return self.pq[self.top];

  def Remove(self):
    if self.top == -1:
      return 'Underflow'
    ret = self.pq[self.top]
    self.top-=1
    
    return ret

```

```python
PQ = Sorted_PQ_Implementation(100)
a=[0,1,2,1,6,9,3,6]
for i in range(len(a)):
  PQ.Insert(a[i])

for i in range(len(a)):
  print(PQ.Remove())
```

```python
#Unsorted
class Unsorted_PQ_Implementation:
  def __init__(self,n):
    self.pq=np.array([0]*n)
    self.top=-1
  def Insert(self,val):
    self.top+=1
    self.pq[self.top]=val
  def Peek(self):
    if self.top == -1:
      return 'Underflow'
    mn=self.pq[0]
    for i in range(self.top+1):
      if self.pq[i] < mn:
        mn=self.pq[i]
    return mn
  def Remove(self):
    if self.top == -1:
      return 'Underflow'
    mn=self.pq[0]
    index=0
    for i in range(self.top+1):
      if self.pq[i] < mn:
        mn=self.pq[i]
        index=i
    for i in range(index,self.top):
      self.pq[i]=self.pq[i+1]
    self.top-=1
    return mn
```

```python
PQ = Unsorted_PQ_Implementation(100)
a=[0,1,2,1,6,9,3,6]
for i in range(len(a)):
  PQ.Insert(a[i])

for i in range(len(a)):
  print(PQ.Remove())
```

```python
def selection(A):
  l = len(A)
  for i in range(l):
    min_id = i
    for j in range(i, l):
      if A[j] < A[min_id]:
        min_id = j
    temp = A[i]
    A[i]= A[min_id]
    A[min_id] = temp
  return A

A= [64, 25, 12, 22,  11]
selection(A)
```

```python
def binary(A, val):
  l = len(A)
  L = 0
  R = l-1
  while L < R:
    M = (L+R)//2
    if A[M] == val:
      return M
    elif A[M] < val: 
      L = M+1
    else: 
      R = M-1
  if A[L] == val:
    return L
  return False 
    
A= [64, 25, 12, 22,  11]
print(selection(A))
print(binary(A, 5))
print(binary(A, 25))
```

```python
 class Node:
   def __init__( self,V,L=None,R=None):
     self.Val=V
     self.L=L
     self.R=R

 class Tree:
   def __init__(self):
     self.root=None
   def Height(self,nd,dp):
     if nd.L==None and nd.R==None:
       return dp
     ldp=0
     rdp=0
     if nd.L!=None:
       ldp=self.Height(nd.L,dp+1)
     if nd.R!=None:
       rdp=self.Height(nd.R,dp+1)
     return max(ldp,rdp)
     
   def level(self,nd,dp,tar):
     if nd==tar:
       return dp
     if nd.L==None and nd.R==None:
       return 0
     ldp=0
     rdp=0
     if nd.L!=None:
       ldp=self.level(nd.L,dp+1,tar)
     if nd.R!=None:
       rdp=self.level(nd.R,dp+1,tar)
     return max(ldp,rdp)
   def preOrder(self,nd):
    print(nd.Val,end="")
    if nd.L!=None:
      self.preOrder(nd.L)
    if nd.R!=None:
      self.preOrder(nd.R)

   def inOrder(self,nd):
    if nd.L!=None:
      self.inOrder(nd.L)
    print(nd.Val,end="")
    if nd.R!=None:
      self.inOrder(nd.R)

   def postOrder(self,nd):
    if nd.L!=None:
      self.postOrder(nd.L)
    if nd.R!=None:
      self.postOrder(nd.R)
    print(nd.Val,end="")


   def identical(self,nd,tempnd):
     if nd==None and tempnd==None:
       return True
     if nd==None or tempnd==None:
       return False
     if nd.Val!=tempnd.Val:
       return False
     
     Left=self.identical(nd.L,tempnd.L)
     Right=self.identical(nd.R,tempnd.R)
     if Left and Right:
       return True
     return False


   def createTree(self,nd):
     if nd==None:
       return nd

     if nd==self.root:

       newTree=Tree()
       newNode=Node(nd.Val)
       newTree.root=newNode

       newNode.L=self.createTree(nd.L)
       newNode.R=self.createTree(nd.R)
       return newTree

     newNode=Node(nd.Val)
     newNode.L=self.createTree(nd.L)
     newNode.R=self.createTree(nd.R)
     return newNode


T = Tree()
B=Node('B')
R=Node('R')
A=Node('A')
C=Node('C')
CC=Node('C')
S=Node('S')
E=Node('E')

T.root=B
B.L=R 
B.R=S 
R.L=A 
R.R=C 
C.L=CC 
S.R=E 


TT = T.createTree(T.root)

T.preOrder(T.root)
print()
TT.preOrder(TT.root)
print()
print(T.identical(T.root,TT.root))

TT.root.R=CC

T.preOrder(T.root)
print()
TT.preOrder(TT.root)
print()

print(T.identical(T.root,TT.root))
```

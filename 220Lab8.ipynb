```python
 class Node:
   def __init__( self,V,L=None,R=None):
     self.Val=V
     self.L=L
     self.R=R

 class Tree:
   def __init__(self):
     self.root=None
   def Height(self,nd,dp):
     if nd.L==None and nd.R==None:
       return dp
     ldp=0
     rdp=0
     if nd.L!=None:
       ldp=self.Height(nd.L,dp+1)
     if nd.R!=None:
       rdp=self.Height(nd.R,dp+1)
     return max(ldp,rdp)

   def level(self,nd,dp,tar):
     if nd==tar:
       return dp
     if nd.L==None and nd.R==None:
       return 0
     ldp=0
     rdp=0
     if nd.L!=None:
       ldp=self.level(nd.L,dp+1,tar)
     if nd.R!=None:
       rdp=self.level(nd.R,dp+1,tar)
     return max(ldp,rdp)
   def preOrder(self,nd):
    print(nd.Val,end="")
    if nd.L!=None:
      self.preOrder(nd.L)
    if nd.R!=None:
      self.preOrder(nd.R)

   def inOrder(self,nd):
    if nd.L!=None:
      self.inOrder(nd.L)
    print(nd.Val,end="")
    if nd.R!=None:
      self.inOrder(nd.R)

   def postOrder(self,nd):
    if nd.L!=None:
      self.postOrder(nd.L)
    if nd.R!=None:
      self.postOrder(nd.R)
    print(nd.Val,end="")


   def identical(self,nd,tempnd):
     if nd==None and tempnd==None:
       return True
     if nd==None or tempnd==None:
       return False
     if nd.Val!=tempnd.Val:
       return False

     Left=self.identical(nd.L,tempnd.L)
     Right=self.identical(nd.R,tempnd.R)
     if Left and Right:
       return True
     return False


   def createTree(self,nd):
     if nd==None:
       return nd

     if nd==self.root:

       newTree=Tree()
       newNode=Node(nd.Val)
       newTree.root=newNode

       newNode.L=self.createTree(nd.L)
       newNode.R=self.createTree(nd.R)
       return newTree

     newNode=Node(nd.Val)
     newNode.L=self.createTree(nd.L)
     newNode.R=self.createTree(nd.R)
     return newNode


T = Tree()
B=Node('B')
R=Node('R')
A=Node('A')
C=Node('C')
CC=Node('C')
S=Node('S')
E=Node('E')

T.root=B
B.L=R
B.R=S
R.L=A
R.R=C
C.L=CC
S.R=E


TT = T.createTree(T.root)

T.preOrder(T.root)
print()
TT.preOrder(TT.root)
print()
print(T.identical(T.root,TT.root))

TT.root.R=CC

T.preOrder(T.root)
print()
TT.preOrder(TT.root)
print()

print(T.identical(T.root,TT.root))
```

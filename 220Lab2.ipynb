```python
TASK1
class Node:
  def __init__(self,val=None):
    self.key=val
    self.ref=None
#TASK2
class MyList:
  def __init__(self, a):
    pre=Node(a[0])
    self.head=pre
    for i in range(1,len(a)):
      new_node=Node(a[i])
      pre.ref=new_node
      pre=new_node
  def showList(self):
    if self.head == None:
      print('Empty List')
      return
    temp=Node()
    temp=self.head
    print(temp.key,end=' ')
    while temp.ref != None:
      temp=temp.ref
      print(temp.key,end=' ')
    print()
  def isEmpty(self):
    if self.head == None:
      return True
    return False
  def clear(self):
    self.head=None
  def insert(self,newElement,index=None):
    if index== None:
      nd=Node(newElement)
      if self.head==None:
        self.head=nd
        return
      pr=self.head
      if pr.key == nd.key:
        print('Key already exists')
        return
      while pr.ref != None:
        pr=pr.ref
        if pr.key == nd.key:
         print('Key already exists')
         return
      pr.ref=nd
    else:
      nd=Node(newElement)
      pre=self.head
      if pre.key == nd.key:
        print('Key already exists')
        return
      while pre.ref != None:
        pre=pre.ref
        if pre.key==nd.key:
          print('Key already exists')
          return
      if index  == 1:   # 1-indexing
       nd.ref=self.head
       self.head=nd
       return
      cnt=1
      pre=None
      cur=self.head
      while cnt<index :
        if cur.ref == None:
          print('Index is not valid')
          return
        cnt+=1
        pre=cur
        cur=cur.ref

      nd.ref=cur
      pre.ref=nd
  def remove(self,deleteKey):
    cur=self.head
    if cur.key==deleteKey:
      self.head=self.head.ref
      return
    pre=None
    while cur.ref !=None:
      pre=cur
      cur=cur.ref
      if cur.key==deleteKey:
        pre.ref=cur.ref
        return

  #Task-3
  def evenNumbers(self):
    cur=self.head
    first=False
    if cur.key%2 == 0:
      print(cur.key,end='')
      first=True
    while cur.ref != None:
      cur=cur.ref
      if cur.key%2 == 0:
        if first:
          print('->',end='')
        print(cur.key,end="")
        first=True
    if first==False:
      print("There is no even number")
    print()
  def isExist(self,val):
    cur=self.head
    if cur.key==val:
      print('True')
      return
    while cur.ref!=None:
      cur=cur.ref
      if cur.key==val:
        print('True')
        return
    print('False')
    return
  def reverse(self):
    if self.head==None or self.head.ref == None:
      return
    pre=self.head
    cur=self.head.ref
    pre.ref=None
    nx=None
    while cur.ref != None:
      nx=cur.ref
      cur.ref=pre
      pre=cur
      cur=nx
    cur.ref=pre
    self.head=cur

  def Sum(self):
    if self.head == None:
      return 0
    sum=0
    nd=self.head
    while nd.ref != None:
      sum+=nd.key
      nd=nd.ref
    sum+=nd.key
    return sum
  def rotate(self,direction,k):
    cnt=1
    cur=self.head
    while cur.ref!=None:
      cnt+=1
      cur=cur.ref
    if direction == 'right':
      k=cnt-k
    cur=self.head
    while cur.ref !=None:
      cur=cur.ref
    cur.ref=self.head
    cnt=0
    cur=self.head
    while True:
      cnt+=1
      if cnt==k:
        self.head=cur.ref
        cur.ref= None
        break
      cur=cur.ref
  def Sort(self):
    if self.head == None:
      return
    cnt=1
    cur=self.head
    while cur.ref!=None:
      cnt+=1
      cur=cur.ref

    if cnt == 1 :
      return
    if cnt == 2:
      F=self.head
      L=self.head.ref
      if F.key>L.key:
        F.ref=None
        L.ref=F
        self.head=L
        return


    for i in range(cnt):
      pre=self.head
      cur=self.head.ref

      if pre.key>cur.key:
        pre.ref=cur.ref
        cur.ref=pre
        self.head=cur

      while cur.ref != None:
        nxt=cur.ref
        if cur.key>nxt.key:
          cur.ref=nxt.ref
          nxt.ref=cur
          pre.ref=nxt
        pre=cur
        cur=nxt
```

```python
Link=MyList([7,0,3,5,2,1,9])
Link.showList()
Link.insert(8,4)
Link.showList()
Link.remove(7)
Link.showList()
Link.remove(2)
Link.showList()
Link.evenNumbers()
Link.insert(2)
Link.insert(10)
Link.evenNumbers()
LL=MyList([1,3,5,7,9,11])
LL.evenNumbers()
LL.isExist(3)
LL.isExist(2)
```

```python


n=int(input())
k=int(input())

score=[None]*n

for i in range(n):
  score[i]=int(input())


a=k-1
for i in range(k,n,1):
  if score[k-1]==score[i]:
   a+=1
  else :
    break

while a>=0 and score[a]<1 :
  a-=1

print(a+1)
```

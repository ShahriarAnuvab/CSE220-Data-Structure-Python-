```python
#Task-1
class Node:
  def __init__(self,key=None,pre=None,nxt=None):
    self.key=key
    self.pre=pre
    self.nxt=nxt
#Task-2
class DoublyList:
  def __init__(self,a):
    prev=Node()
    for i in range(len(a)):
      cur=Node(a[i])
      if i == 0:
        self.head=cur
      else:
        prev.nxt=cur
        cur.pre=prev
      prev=cur
    self.head.pre=prev
    prev.nxt=self.head
  def showList(self):
    cur=self.head
    if cur==None:
      print('Empty List')
    while True:
      print(cur.key,end=' ')
      if cur.nxt == self.head:
        break
      cur=cur.nxt
    print()
  def insert(self,newElement,index=None):
    newNode=Node(newElement)
    if index == None:
      cur=self.head
      if cur == None:
        self.head=newNode
      else:
        while True:
          if cur.key==newNode.key:
            print('Key already exists')
            return
          if cur.nxt==self.head:
            break
          cur=cur.nxt
        hd=self.head
        lst=self.head.pre
        self.head.pre=newNode
        lst.nxt=newNode
        newNode.nxt=self.head
        newNode.pre=lst
    else:
      cnt=0
      cur=self.head
      while True:
        cnt+=1
        if cur.key == newNode.key:
          print('Key already exists')
          return
        if cur.nxt == self.head:
          break
        cur=cur.nxt
      if cnt < index:
        print('Invalid Index')
        return
      cnt=0
      cur=self.head
      while True:
        cnt+=1
        if cnt==index:
          previous = cur.pre
          Next= cur
          previous.nxt=newNode
          Next.pre=newNode
          newNode.pre=previous
          newNode.nxt=Next
          if index == 1:
            self.head=newNode
          return
        cur=cur.nxt
  def remove(self,index):
    cnt=0
    cur=self.head
    while True:
      cnt+=1
      if cnt==index:
        cur.pre.nxt=cur.nxt
        cur.nxt.pre=cur.pre
        #previous=cur.pre
        #Next=cur.nxt
        #previous.nxt=Next
        #Next.pre=previous
        if index == 1:
          self.head=cur.nxt
        return
      if cur.nxt == self.head:
        break
      cur=cur.nxt
    print('Invalid Index')
  def removeKey(self,key):
    cur=self.head
    while True:
      if cur.key == key:
        cur.pre.nxt=cur.nxt
        cur.nxt.pre=cur.pre
        if cur==self.head:
          self.head=cur.nxt
        return cur.key
      if cur.nxt == self.head:
        return 'Key does not exists'
      cur=cur.nxt
```
